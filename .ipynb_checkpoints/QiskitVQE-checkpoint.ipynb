# Vacuum State Prep

## Construct the Hamiltonian

#### Constants

```python
g = math.sqrt(16/3)
m = 1
```

#### Helper Functions

```python
def pauli_op(op,loc,nq,cons = 1):
    operator = SparsePauliOp.from_sparse_list([(op, [loc], cons)], num_qubits = nq)
    return operator
```

```python
def projection(which, loc, nq):
    if which == 0:
        operator = SparsePauliOp.from_sparse_list([("Z", [loc], 1), ("I", [loc], 1)], num_qubits=nq)/2
    elif which == 1:
        operator = SparsePauliOp.from_sparse_list([("Z", [loc], -1), ("I", [loc], 1)], num_qubits=nq)/2
    else:
        raise Exception("Type of projection operator not supported. Choose 0 or 1.")
    return operator
```

#### Kinetic Hamiltonian

```python
def Hkin(nq):
    op1 = SparsePauliOp.from_sparse_list([("I", [0], 0)], num_qubits = nq)
    op2 = SparsePauliOp.from_sparse_list([("I", [0], 0)], num_qubits = nq)

    for i in range(nq):
        op1 += projection(0,i,nq).dot(pauli_op("X",(i+1)%nq,nq)).dot(projection(0,(i+2)%nq,nq))
        op2 += projection(1,i,nq).dot(pauli_op("X",(i+1)%nq,nq)).dot(projection(1,(i+2)%nq,nq))
        
    return (math.sqrt(2)*op1/2 + op2/math.sqrt(2)/2).simplify()
```

#### Mass Hamiltonian

```python
def Hm(nq,m):
    op = SparsePauliOp.from_sparse_list([("I", [0], 0)], num_qubits = nq)

    for i in range(nq):
        op += projection(0,i,nq).dot(projection(1,(i+1)%nq,nq)) + projection(1,i,nq).dot(projection(0,(i+1)%nq,nq))
        
    return (m*op).simplify()
```

#### Electric Hamiltonian

```python
def He(nq,g):
    op = SparsePauliOp.from_sparse_list([("I", [0], 0)], num_qubits = nq)
    
    for i in range(nq):
        op += projection(1,i,nq) 
        
    return (3*g**2/8*op).simplify()
```

```python
def He_persite(nq,g,site):
    return 3 * g**2/8 * projection(1,site,nq) 
```

#### Full Hamiltonian with mass m and gauge coupling g

```python
def Hfull(nq,m,g):
    return (Hkin(nq)+He(nq,g)+Hm(nq,m)).simplify()
```

## Evolution Operators

```python
def two_qubit_gates_layers(ZY_ang,YZ_ang,nq,circ):
    n_sets = int(nq/2)
    
    for j in range(n_sets): # (1,2),(3,4),(5,6) 
        # Do ZY and YZ 
        
        a,b = 2*j,(2*j+1)%nq
    
        circ.sdg(a)
        circ.h(a)
        circ.cx(a,b)
        circ.rz(YZ_ang,b)
        circ.cx(a,b)
        circ.h(a)
        circ.s(a)
    
        circ.sdg(b)
        circ.h(b)
        circ.cx(a,b)
        circ.rz(ZY_ang,b)
        circ.cx(a,b)
        circ.h(b)
        circ.s(b)
        
    for jj in range(n_sets): # (2,3),(4,5),(6,1) 
        c,d = (2*jj+1)%nq,(2*jj+2)%nq

        circ.sdg(d)
        circ.h(d)
        circ.cx(c,d)
        circ.rz(YZ_ang,d)
        circ.cx(c,d)
        circ.h(d)
        circ.s(d)

        circ.sdg(c)
        circ.h(c)
        circ.cx(c,d)
        circ.rz(ZY_ang,d)
        circ.cx(c,d)
        circ.h(c)
        circ.s(c)
```

```python
def three_qubit_gates_layers(angle,nq,circ):
    n_groups = int(nq/3) 
    for m in range(n_groups): # (1,2,3),(4,5,6),(7,8,9) #ZYZ
        aa,bb,cc = (3*m)%nq,(3*m+1)%nq,(3*m+2)%nq
        circ.sdg(bb)
        circ.h(bb)
        circ.cx(aa,bb)
        circ.cx(bb,cc)
        circ.rz(angle,cc)
        circ.cx(bb,cc)
        circ.cx(aa,bb)
        circ.h(bb)
        circ.s(bb)
    for mm in range(n_groups): # (2,3,4),(5,6,7),(8,9,1)
        aa,bb,cc = (3*mm+1)%nq,(3*mm+2)%nq,(3*mm+3)%nq
        circ.sdg(bb)
        circ.h(bb)
        circ.cx(aa,bb)
        circ.cx(bb,cc)
        circ.rz(angle,cc)
        circ.cx(bb,cc)
        circ.cx(aa,bb)
        circ.h(bb)
        circ.s(bb)
    for mmm in range(n_groups): # (3,4,5),(6,7,8),(9,1,2)
        aa,bb,cc = (3*mmm+2)%nq,(3*mmm+3)%nq,(3*mmm+4)%nq
        
        circ.sdg(bb)
        circ.h(bb)
        circ.cx(aa,bb)
        circ.cx(bb,cc)
        circ.rz(angle,cc)
        circ.cx(bb,cc)
        circ.cx(aa,bb)
        circ.h(bb)
        circ.s(bb)
```

```python
# define my_form

def one_layer(nq,thetas): #thetas are the parameters
    circ = QuantumCircuit(nq)
    it = iter(thetas)

    '''one qubit gates layer'''
    # all one qubit gates have the same angle
    one_qubit_angle = next(it)
    for i in range(nq):
        circ.ry(one_qubit_angle, i)

    '''two qubit gates layer'''
    # ZY two qubit gates have the same angle; YZ two qubit gates have the same angle;
    ZY_angle = next(it)
    YZ_angle = next(it)
    two_qubit_gates_layers(ZY_angle,YZ_angle,nq,circ)

    ''' three qubit gates layer '''
    tri_angle = next(it)
    three_qubit_gates_layers(tri_angle,nq,circ)

    '''two qubit gates layer'''
    ZY_angle = next(it)
    YZ_angle = next(it)
    two_qubit_gates_layers(ZY_angle,YZ_angle,nq,circ)
    
    return circ
```
