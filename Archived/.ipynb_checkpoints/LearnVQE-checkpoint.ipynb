## Functions

```python
def Reflect(cir = None):
    if cir == None: cir = QuantumCircuit(nQ)
    cir.h(nQ-1)
    for i in range(nQ-1):
        cir.cx(nQ-1, nQ-2-i)
    return cir
```

```python
def init_circuit_params(angles_param,nq = nQ,cir = None):
    if cir == None: cir = QuantumCircuit(nq)
    site = [nq-2-x for x in range(0, nq-1)]
    it = iter(angles_param)
    cir.ry(next(it), site[0])
    for i in range(1,nq-1):
        for k in range(0, 2**i):
            cir.ry(next(it),site[i])
            # if k%2 == 0:
            #      cir.cx(site[i-1],site[i])
            for l in range(0,i): 
                if np.mod(k+1, 2**(i-l-1)) == 0:
                    cir.cnot(site[l], site[i])
                    break    
    Reflect(cir)
    return cir
```

```python
def padDict(dicta):
    ordered_prob = []
    for i in range(2**nQ):
        if i in dicta.keys(): 
            ordered_prob.append(dicta[i])
        else: ordered_prob.append(0)
    return ordered_prob 
```

```python
StateParams = ParameterVector('theta', length=2**(nQ-1)-1)

qr = QuantumRegister(5, name="q")
cr = ClassicalRegister(5, name="c")
my_form = QuantumCircuit(qr,cr)
my_form = init_circuit_params(StateParams)
my_form.measure_all()
```

### One way

```python
sampler = Sampler()

target_distr = State5g # an np array

def Overlap(State1, State2):
    return (np.vdot(np.array(State1),np.array(State2)))**2

def objective_function(params):
    
    result = sampler.run(circuits=my_form, parameter_values=params).result()
    
    output_distr = result.quasi_dists[0]
    cost = 1-Overlap(target_distr, padDict(output_distr))
    return cost
```

```python
optimizer = COBYLA(maxiter=5000, tol=0.0001)
# Create the initial parameters (noting that our
# single qubit variational form has 3 parameters)
initial_point = np.random.rand(2**(nQ-1)-1)

result = optimizer.minimize(fun=objective_function, x0=initial_point)

# Obtain the output distribution using the final parameters
output_distr = (
    sampler.run(circuits=my_form, parameter_values=result.x)
    .result()
    .quasi_dists[0]
)

print("Parameters Found:", result.x)
print("Target Distribution:", target_distr)
print("Obtained Distribution:", output_distr)
print("Cost:", objective_function(result.x))
```
